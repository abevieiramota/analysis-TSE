```python
import pandas as pd
%matplotlib inline
```

```python
FILEPATH = "../scripts/estatistica/sead/odsele/prestacao_contas/prestacao_final_2012/candidato/CE/ReceitasCandidatos.txt"
```

```python
df = pd.read_csv(FILEPATH, sep=";", decimal=",")
```

```python
def d(c):
    df[[c, 'Valor receita']].groupby(c).sum().sort('Valor receita', ascending=False)[:20].plot(kind='barh', title=c)
```
