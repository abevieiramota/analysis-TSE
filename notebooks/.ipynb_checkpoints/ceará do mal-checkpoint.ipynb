```python
import pandas as pd
%matplotlib inline
```

```python
SEP = ';'
ENCODING = "iso-8859-1"
```

```python
CAND_CE = "../scripts/estatistica/sead/odsele/consulta_cand/consulta_cand_2014/consulta_cand_2014_CE.txt"
BEM_CE = "../scripts/estatistica/sead/odsele/bem_candidato/bem_candidato_2014/bem_candidato_2014_CE.txt"

candCE = pd.read_csv(CAND_CE, sep=SEP, encoding=ENCODING, usecols=[9, 10, 11, 12, 13, 14, 15], names=['cargo', 'nome', 'numero', 'oque', 'nomenome', 'num', 'status'])
bemCE = pd.read_csv(BEM_CE, sep=SEP, encoding=ENCODING, usecols=[5, 6, 7, 8, 9, 10, 11], names=['numero', 'seila', 'tipo', 'oqueeh', 'valor', 'data', 'hora'])
```

```python
juntinhos = bemCE.groupby('numero').sum().merge(candCE, left_index=True, right_on='numero')
```
