```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('../20140805_expulsoes.tsv', sep=r'\t', converters={5:lambda x: x.lower()})
df.count()
```

```python
qtd_por_orgao = df[[df.columns[3]]].groupby(df.columns[3]).count()
```

```python
qtd_por_orgao.sort(columns=df.columns[3], ascending=False)[:20].plot(kind='barh', figsize=(10,10))
```

```python
df.columns
```

```python
uf = df.columns[4]
df[[uf]].groupby(uf).count().sort(columns=uf, ascending=False).plot(kind='barh', figsize=(8,8))
```

```python
cargo = df.columns[5]
df[[cargo]].groupby(cargo).count().sort(columns=cargo, ascending=False)[:10].plot(kind='barh', figsize=(8,8))
```

```python
data = df.columns[9]
df[[data]].applymap(lambda x: x.split('/')[2]).groupby(data).count().sort(columns=data, ascending=False).plot(kind='barh', figsize=(8,8))
```
